```python
import numpy as np

# Предположим, что 'sorted_feature2' - это массив NumPy или список значений
# Если у вас DataFrame, то можно получить столбец 'sorted_feature2' так: sorted_feature2 = df['sorted_feature2'].values

# Вычисляем четвёртый центральный момент
fourth_central_moment = np.mean((feature2_column - np.mean(feature2_column))**4)

# Выводим результат
print("Четвёртый центральный момент:", fourth_central_moment)
```

```python


# Рассчитываем выборочное стандартное отклонение в четвертой степени
sample_std_deviation_fourth_power = np.std(feature2_column)**4
sample_std_deviation = np.std(feature2_column)
# Замените 'confidence_interval' на вашу переменную с доверительным интервалом
confidence_interval = 0.95

# Рассчитываем t
t = (fourth_central_moment / (sample_std_deviation_fourth_power * (1 - confidence_interval)))**(1/4)
print("Значение t:", t)
```

```python


n = len(feature2_column)  # Количество элементов в ряде

# Рассчитываем медиану (Q2)
median_index = n // 2
Q2 = (feature2_column[median_index] + feature2_column[median_index - 1]) / 2 if n % 2 == 0 else feature2_column[median_index]
Q  = sum(feature2_column) / len(feature2_column)

# Рассчитываем верхнюю и нижнюю границы предельных значений
upper_limit = Q + t * (sample_std_deviation)
lower_limit = Q - t * (sample_std_deviation)
print("Медиана:", Q2)
print("Среднее арифметическое:", Q)
print("выборочное стандартное отклонение:", sample_std_deviation)
print("Верхняя граница:", upper_limit)
print("Нижняя граница:", lower_limit)
```

```python
outliers = [x for x in feature2_column if x < lower_limit or x > upper_limit]

if outliers:
    print("Есть выбросы за пределами границ:")
    print([round(x, 2) for x in outliers])
else:
    print("Выбросов за пределами границ не обнаружено.")
```

```python
filtered_data = [x for x in feature2_column if lower_limit <= x <= upper_limit]
```

```python
mean = np.mean(filtered_data)  # Среднее арифметическое
squared_diff = [(x - mean) ** 2 for x in filtered_data]  # Квадраты разностей от среднего
variance = sum(squared_diff) / (len(filtered_data) - 1)  # Несмещенная оценка дисперсии
print("Несмещенная оценка дисперсии:", variance)
```

```python
std_deviation = np.sqrt(variance)
print("Стандартное отклонение:", std_deviation)
```

```python
n = len(filtered_data)  # Размер выборки
third_central_moment = sum([(x - mean) ** 3 for x in filtered_data]) / n
print("Третий центральный момент:", third_central_moment)
```

```python
skewness = third_central_moment / (variance ** (3 / 2))
print("Безразмерный коэффициент асимметрии:", skewness)
```

```python
skewness_variance = (6 * (n - 2)) / ((n + 1) * (n + 3))**(1/2)
print("Оценка коэффициента асимметрии:", skewness_variance)
```

```python
# Вычисляем эксцесс
excess = fourth_central_moment / (variance**2)

print("Эксцесс:", excess)
```

```python
kurtosis = 1/ (excess)**(1/2)

print("Контрэксцесс:", kurtosis)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Определите количество интервалов
num_intervals = sterges

# Рассчитайте минимальное и максимальное значение в данных
min_value = np.min(filtered_data)
max_value = np.max(filtered_data)

# Рассчитайте ширину интервала
interval_width = (max_value - min_value) / num_intervals

# Рассчитайте границы интервалов
interval_boundaries = np.arange(min_value, max_value + interval_width, interval_width)

# Рассчитайте гистограмму, включая количество отсчетов в каждом интервале
hist, bins = np.histogram(filtered_data, bins=interval_boundaries)

# Создайте гистограмму
plt.hist(filtered_data, bins=interval_boundaries, edgecolor='k', alpha=0.7)
plt.title('Гистограмма данных')
plt.xlabel('Значения')
plt.ylabel('Частота')
plt.grid(True)

# Выведите количество данных в каждом интервале
for i, count in enumerate(hist):
    print(f"Интервал {i + 1}: {count} данных")

# Отобразите гистограмму
plt.show()
```

```python
# Вычислите вероятности для каждого интервала
probabilities = hist / np.sum(hist)

# Вычислите оценку функции распределения F(Q)
cdf = np.cumsum(probabilities)

# Постройте график оценки функции распределения
plt.plot(bins[:-1], cdf, marker='o', linestyle='-')
plt.xlabel('Q')
plt.ylabel('F(Q)')
plt.title('Оценка функции распределения')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.stats import kstest
import matplotlib.pyplot as plt


# Теоретическое распределение (PDF) - замените его на ваше теоретическое распределение
a = -0.25
b = 0
c = 0.8
d = 1.0

x = np.linspace(a - 0.5, d + 0.5, 1000)
pdf = []

for val in x:
    if a <= val < b:
        pdf.append((2 / (d + c - a - b)) * (val - a) / (b - a))
    elif b <= val < c:
        pdf.append(2 / (d + c - a - b))
    elif c <= val <= d:
        pdf.append((2 / (d + c - a - b)) * (d - val) / (d - c))
    else:
        pdf.append(0)

# Создайте экспериментальную функцию распределения (CDF) на основе ваших данных
def empirical_cdf(data, x):
    count = [np.sum(data <= val) for val in x]
    return np.array(count) / len(data)

# Вычислите экспериментальную функцию распределения (CDF)
empirical_cdf_values = empirical_cdf(filtered_data, x)

# Вычислите теоретическую функцию распределения (CDF)
theoretical_cdf_values = np.cumsum(pdf) / np.sum(pdf)

# Постройте графики экспериментальной и теоретической CDF
plt.plot(x, empirical_cdf_values, label='Экспериментальная CDF')
plt.plot(x, theoretical_cdf_values, label='Теоретическая CDF')
plt.xlabel('Значение')
plt.ylabel('Функция распределения вероятности')
plt.legend()
plt.grid(True)
plt.title('Сравнение экспериментальной и теоретической CDF')
plt.show()

# Примените Критерий согласия Колмогорова
ks_statistic, ks_p_value = kstest(filtered_data, cdf=theoretical_cdf_values)

# Выведите результаты
print(f"Значение статистики Колмогорова: {ks_statistic}")
print(f"p-значение: {ks_p_value}")

# Оцените результаты
alpha = 0.05  # Уровень значимости (можно изменить)
if ks_p_value < alpha:
    print("Гипотеза о согласии отвергается")
else:
    print("Гипотеза о согласии принимается")
```
